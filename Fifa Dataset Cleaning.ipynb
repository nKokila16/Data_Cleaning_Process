```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading dataset
path='fifa_23_data.csv'
df=pd.read_csv(path)
```

```python
#reading dataset by displaying only first rows
df.head()
```

```python
#exploring dataset
print(df.columns)
print(df.info())
```

```python
print("Summary Statistics:")
print(df.describe())
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
# Check for duplicate rows in the entire DataFrame
duplicate_rows = df[df.duplicated()]

# Display duplicate rows
print("Duplicate Rows:")
print(duplicate_rows)

```

```python
# Remove all duplicate rows from the DataFrame
df_fifa_no_duplicates = df.drop_duplicates()

# Display the first few rows of the DataFrame without duplicates
print("DataFrame without Duplicates:")
print(df_fifa_no_duplicates.head())
```

```python
df.shape
```

```python
df.sort_values(by="Overall",ascending=False).head(5)
```

```python
df.sort_values(by=["Overall","Age"],ascending=True).head(5)
```

```python
#grouping the data
group_by_nationality = df.groupby('Nationality')
avg_players_by_age = group_by_nationality["Age"].mean().head()
print(avg_players_by_age)
```

```python
import matplotlib.pyplot as plt

# Plot a bar chart of average age by nationality
avg_players_by_age.plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title('Average Age of Soccer Players by Nationality')
plt.xlabel('Nationality')
plt.ylabel('Average Age')
plt.show()
```

```python
df['Nationality'].unique()
```

```python
# Set 'Nationality' as the index
df_fifa_indexed = df.set_index('Nationality')

# Grouping by the index
grouped_by_index = df_fifa_indexed.groupby(level='Nationality')

# Applying an aggregation function, for example, calculating the average age in each nationality group
average_age_by_index = grouped_by_index['Age'].mean()

# Displaying the result
print(average_age_by_index)
```

```python
# Grouping by multiple columns, for example, 'Nationality' and 'Club Name'
grouped_by_nationality_club = df.groupby(['Nationality', 'Club Name'])

# Applying an aggregation function, for example, calculating the average overall rating in each nationality-club group
average_rating_by_nationality_club = grouped_by_nationality_club['Overall'].mean()

# Displaying the result
print(average_rating_by_nationality_club)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Overall'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Player Ratings in FIFA')
plt.xlabel('Player Rating (Overall)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Overall'], alpha=0.5, color='orange')
plt.title('Scatter Plot of Player Age vs. Overall Rating')
plt.xlabel('Age')
plt.ylabel('Overall Rating')
plt.grid(True)
plt.show()
```

```python
top_nationalities = df['Nationality'].value_counts().head(10)
plt.figure(figsize=(12, 6))
top_nationalities.plot(kind='bar', color='green')
plt.title('Top 10 Nationalities in FIFA Dataset')
plt.xlabel('Nationality')
plt.ylabel('Number of Players')
plt.xticks(rotation=45, ha='right')
plt.show()
```
